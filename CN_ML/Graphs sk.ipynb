```python
# plt.plot(x,y,color="blue",marker="+")
x=np.arange(0,5,0.5)
y=x**2
y2=x**3
plt.plot(x,y,color="red",marker="^",linewidth= 3 ,label="x^2")
plt.plot(x,y2,color="green",marker="o",linewidth= 3 , label="x^3")
plt.ylabel("y")
plt.xlabel('x')
plt.legend()
plt.text(3,26,"hare krishna",fontsize=12)
plt.grid()
```

```python
#pie charts
a=[2,3,7]
colors=["orange","blue","yellow"]
labels=['class A',"class B","class c"]
explode=[0.2,0.5,0]
plt.title("Each class score",fontsize=15)
plt.pie(a,colors=colors,explode=explode ,labels=labels,autopct='%.1f%%',counterclock=False,startangle=100)
plt.show()
```
